## Verileri etiketleyelim

```python
import os
import cv2
```

```python
image_names = os.listdir("./sub_ds")
```

```python
labels = []

for image_name in image_names:
    im = cv2.imread("./sub_ds/"+image_name)
    
    im = cv2.resize(im,(160,160))
    cv2.imshow(image_name,im)
    key = cv2.waitKey(0)
    if key == ord("k"):
        labels.append(0)
        cv2.destroyAllWindows()
        continue
    elif key ==ord("h"):
        labels.append(1)
        cv2.destroyAllWindows()
        continue
    elif key == ord("q"):
        cv2.destroyAllWindows()
        break
    else:
        cv2.destroyAllWindows()
        continue
```

```python
len(labels)
```

```python
len(image_names)
```

```python
labels
```

```python
images= []
for image_name in image_names:
    im = cv2.imread("./sub_ds/"+image_name)
    images.append(im)
```

```python
len(images)
```

```python
import numpy as np
images = np.array(images)
labels = np.array(labels)
```

## Modeli Eğitelim

```python
from tensorflow.keras.layers import Conv2D,MaxPooling2D,Dense,Input,Flatten
from tensorflow.keras.models import Sequential
```

```python
model = Sequential()
model.add(Input(shape=(40,40,3)))
model.add(Conv2D(25,(3,3),))
model.add(Conv2D(25,(3,3),))
model.add(Conv2D(25,(5,5),))

model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(25,(5,5),))

model.add(Flatten())
model.add(Dense(1,activation='sigmoid'))
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
model.fit(x=images,y=labels,epochs=100)
```

```python
f = open("sub_ds_labels.txt","w")
for lbl in labels:
    f.write(str(lbl)+",")
```
