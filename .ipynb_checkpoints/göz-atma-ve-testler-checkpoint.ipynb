```python
tags ={}

for sub_file in tqdm(sub_files):
    general_videos_path = f"./{main_files[0]}/{sub_file}/"
    general_txts_path = f"./{main_files[1]}/{sub_file}/"
    
    video_files_path = os.listdir(general_txts_path)
    
    for video_file_path in tqdm(video_files_path):
        video_path = general_videos_path + video_file_path + "/video.mp4"
        txt_path = general_txts_path + video_file_path + "/annotations.txt"
        print("İşleniyor:")
        print(video_path)
        print(txt_path)
        print()
       
    
        # txt file
        f = open(txt_path,"r")
        lines = f.readlines()
        for line in lines:
            words = line.split(" ")
            x1 = words[1]
            y1 = words[2]
            x2 = words[3]
            y2 = words[4]
            lbltext = words[9] 
    
            try:
                tags[lbltext] += 1
            except:
                tags[lbltext] = 1
        
    clear_output()

pd.DataFrame(tags.values(),index=tags.keys())
```

```python

color = (0,123,255)
frame_count = 0
cap = cv2.VideoCapture(video_path)

while(cap.isOpened()):
    ret, frame = cap.read()
    frame_ = cv2.resize(frame,(frame_shape[1],frame_shape[0]))
    
    if ret == True:
        
        
        # objects
        boxes = bboxes[frame_count] 
        for box in boxes:
            x1 = box[0]
            y1 = box[1]
            x2 = box[2]
            y2 = box[3]
            lbltext = box[4]
            id_ = box[5]
            lost = box[6]
            occuled = box[7]
            generated = box[8]
            
            if x1 > detect_area_bbox[0][1] and x2 < detect_area_bbox[1][1] and y1 > detect_area_bbox[0][0] and y2 < detect_area_bbox[1][0]:
                
                if lost == 0 and occuled == 0:
                    
                    ## dataset classic label and boxes
                    #cv2.rectangle(frame_,(x1,y1),(x2,y2),color)
                    #cv2.putText(frame_,lbltext+str(id_),(x1,y1),cv2.FONT_HERSHEY_SIMPLEX,1,color,1,cv2.LINE_AA)
                    
                    # with our labels
                    cv2.putText(frame_,str(target_labels[lbltext]),(x1,y1),cv2.FONT_HERSHEY_SIMPLEX,1,color,1,cv2.LINE_AA)
                    
                    # boxes with yolo format
                    center_x = (x2+x1) / 2
                    center_y = (y2+y1) / 2
                    width = x2-x1
                    height = y2-y1
                    cv2.line(frame_,(int(center_x-width/2),int(center_y)),(int(center_x+width/2),int(center_y)),color)
                    cv2.line(frame_,(int(center_x),int(center_y-height/2)),(int(center_x),int(center_y+height/2)),color)
                    
                    # location ratios in yolo format
                    max_y = frame_shape[0]
                    max_x = frame_shape[1]
                    cx = center_x / max_x
                    cy = center_y / max_y
                    w = width / max_x
                    h = height / max_y
                    
                    # locating on yolo format
                    #cv2.putText(frame_,f"{cx:.2f} {cy:.2f} {w:.2f} {h:.2f}",(x1,y1),cv2.FONT_HERSHEY_SIMPLEX,1,color,1,cv2.LINE_AA)
                    
        cv2.rectangle(frame_,detect_area_bbox[0][::-1],detect_area_bbox[1][::-1],color)
        frame_ = cv2.resize(frame_,(1200,1000))
        cv2.imshow('Frame',frame_)
        frame_count +=1

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else: 
        break

cap.release()
cv2.destroyAllWindows()
```

```python
def prepare_labels(reference_im_path,txt_path):
    bboxes = {}
    
    padding = 50
    frame_shape = cv2.imread(reference_im_path).shape[:2]
    detect_area_bbox = [(padding,padding),(frame_shape[0]-padding,frame_shape[1]-padding)]

    file = open(txt_path,"r")
    for line in file.readlines():
        words = line.split(" ")

        try:
            bboxes[int(words[5])].append([
                int(words[1]),
                int(words[2]),
                int(words[3]),
                int(words[4]),
                words[9][1:-2],
                int(words[0]),
                int(words[6]),
                int(words[7]),
                int(words[8]),
            ])
        except:
            bboxes[int(words[5])] = [[
                int(words[1]),
                int(words[2]),
                int(words[3]),
                int(words[4]),
                words[9][1:-2],
                int(words[0]),
                int(words[6]),
                int(words[7]),
                int(words[8]),
            ]]
    return bboxes
```

```python

def picking(video_path,reference_im_path,anno_path):
    color = (0,123,255)
    
    
    
    bboxes =prepare_labels(refence_im_path,anno_path)

    cap = cv2.VideoCapture(video_path)
    reference_shape = cv2.imread(refence_im_path).shape    

    frame_count = 0
    while(cap.isOpened()):
        ret, frame = cap.read()
        frame_ = cv2.resize(frame,(reference_shape[1],reference_shape[0]))

        if ret == True:

            lines = []
            # objects
            boxes = bboxes[frame_count] 
            for box in boxes:
                x1 = box[0]
                y1 = box[1]
                x2 = box[2]
                y2 = box[3]
                lbltext = box[4]
                id_ = box[5]
                lost = box[6]
                occuled = box[7]
                generated = box[8]

                if x1 > detect_area_bbox[0][1] and x2 < detect_area_bbox[1][1] and y1 > detect_area_bbox[0][0] and y2 < detect_area_bbox[1][0]:

                    if lost == 0 and occuled == 0:

                        # boxes with yolo format
                        center_x = (x2+x1) / 2
                        center_y = (y2+y1) / 2
                        width = x2-x1
                        height = y2-y1
                        cv2.line(frame_,(int(center_x-width/2),int(center_y)),(int(center_x+width/2),int(center_y)),color)
                        cv2.line(frame_,(int(center_x),int(center_y-height/2)),(int(center_x),int(center_y+height/2)),color)

                        # location ratios in yolo format
                        max_y = frame_shape[0]
                        max_x = frame_shape[1]

                        label = target_labels[lbltext]
                        cx = center_x / max_x
                        cy = center_y / max_y
                        w = width / max_x
                        h = height / max_y

                        lines.append(f"{label} {cx} {cy} {w} {h}")
                        # locating on yolo format
                        #cv2.putText(frame_,f"{cx:.2f} {cy:.2f} {w:.2f} {h:.2f}",(x1,y1),cv2.FONT_HERSHEY_SIMPLEX,1,color,1,cv2.LINE_AA)

            cv2.rectangle(frame_,detect_area_bbox[0][::-1],detect_area_bbox[1][::-1],color)
            frame_ = cv2.resize(frame_,(1200,1000))
            cv2.imshow('Frame',frame_)
            frame_count +=1

            if cv2.waitKey(1) & 0xFF == ord('q'):
                break
        else: 
            break

    cap.release()
    cv2.destroyAllWindows()
```
