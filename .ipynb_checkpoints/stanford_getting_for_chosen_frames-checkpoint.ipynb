```python
def path_sequence(root,*directories):
    seq = root
    for directory in directories:
        seq += f"/{directory}"
    return seq
```

1   Track ID. All rows with the same ID belong to the same path.
2   xmin. The top left x-coordinate of the bounding box.
3   ymin. The top left y-coordinate of the bounding box.
4   xmax. The bottom right x-coordinate of the bounding box.
5   ymax. The bottom right y-coordinate of the bounding box.
6   frame. The frame that this annotation represents.
7   lost. If 1, the annotation is outside of the view screen.
8   occluded. If 1, the annotation is occluded.
9   generated. If 1, the annotation was automatically interpolated.
10  label. The label for this annotation, enclosed in quotation marks.

```python

target_labels = {
    '"Pedestrian"\n' : 11,
    '"Biker"\n': 1,
    '"Skater"\n' : 4,
    '"Cart"\n': 4,
    '"Car"\n': 0 ,
    '"Bus"\n' : 2
}

target_counts = {
    '"Pedestrian"\n' : 0,
    '"Biker"\n': 0,
    '"Skater"\n' : 0,
    '"Cart"\n': 0,
    '"Car"\n': 0 ,
    '"Bus"\n' : 0
}

def extract_line(line,ref_shape_x,ref_shape_y):
    words =line.split(" ")
    
    track_id = int(words[0])
    x1 = int(words[1])
    y1 = int(words[2])
    x2 = int(words[3])
    y2 = int(words[4])
    frame_idx = int(words[5])
    lost = int(words[6])
    occlueded = int(words[7])
    generated = int(words[8])
    label = target_labels[words[9]]
    
    target_counts[words[9]] +=1
    
    center_x = int((x1+x2)/2)
    center_y = int((y1+y2)/2)
    width = x2-x1
    height = y2-y1
    
    norm_center_x = center_x / ref_shape_x
    norm_center_y = center_y / ref_shape_y
    norm_width = width / ref_shape_x
    norm_height = height / ref_shape_y    
    
    return [label,norm_center_x,norm_center_y,norm_width,norm_height,frame_idx]
```

```python
frame_interval = 30
save_new_labels = True
```

```python
root = "./video"

labels_root = "./annotations"
locations_list =  os.listdir(root)
for location in tqdm(locations_list):
    video_files_list = os.listdir(path_sequence(root,location))
    
    frame_indexes = []
    
    frame_tags = {}
        
    for video_file in tqdm(video_files_list):
        
        print(f"işleniyor: {labels_root} {location} {video_file} annotations")

        video_file_path = path_sequence(root,location,video_file)
        new_tags_path = path_sequence(video_file_path,"labels")
        
        if os.path.exists(new_tags_path) == False:   
            os.mkdir(new_tags_path)
            
        # annoya git ve indextekileri al
        labels_file_path = path_sequence(labels_root,location,video_file,"annotations.txt")
        ref_im_path = path_sequence(labels_root,location,video_file,"reference.jpg")
        ref_im = cv2.imread(ref_im_path)
        
        f = open(labels_file_path,"r")
        lines = f.readlines()
        for line in lines:
            newline = extract_line(line,ref_im.shape[1],ref_im.shape[0])
            if newline[5] % frame_interval == 0:
                if newline[5] in frame_tags and frame_tags[newline[5]] is not None:
                    frame_tags[newline[5]].append(newline)
                else:
                    frame_tags[newline[5]] = [newline]
        
        if save_new_labels:
            for frame_idx in frame_tags.keys():
                f= open(path_sequence(root,location,video_file,"labels",f"{frame_idx}.txt"),"w")
                for tag in frame_tags[frame_idx]:
                    tag_strings = [str(item) for item in tag[:5]]
                    linestring = " ".join(tag_strings)
                    linestring += "\n"
                    f.write(linestring)
                f.close()  # Dosyayı kapatmayı unutmayın
        #print(frame_tags)
```
